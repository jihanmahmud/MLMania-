```python
import pandas as pd
import requests
```

```python
response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=1af56306157684c07649fa74501fe09b")
df= pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'popularity', 'release_date', 'vote_average', 'vote_count']]
```

```python
for i in range(1,471):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=1af56306157684c07649fa74501fe09b&page={}'.format(i))
    temp_df= pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'popularity', 'release_date', 'vote_average', 'vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True)
```
