```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('train.csv')

# Print column names
print("Columns:", df.columns)

# Find the correct column name (case-insensitive)
sex_column = next((col for col in df.columns if col.lower() == 'sex'), None)

if sex_column:
    # Create distribution plot
    sns.histplot(data=df, x=sex_column, kde=True)
    plt.title(f"Distribution of {sex_column}")
    plt.show()
else:
    print("No 'sex' column found. Available columns are:", df.columns)
```
